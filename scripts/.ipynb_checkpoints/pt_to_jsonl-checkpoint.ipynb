```python
PT_FILE = '/home/ubuntu/training_data/20230224-full-crawl-ds-filter-corrected/notebook.test.input_ids.pt'
# PT_FILE = '/home/ubuntu/training_data/20230208/processed/python.test.input_ids.pt'
base_model_name = 'Salesforce/codegen-2B-mono'
```

```python
import torch
data = torch.load(PT_FILE)
```

```python
data.shape
```

```python
from transformers import (
    AutoConfig, AutoModelForCausalLM, AutoTokenizer, CodeGenForCausalLM,
    EarlyStoppingCallback, Trainer, TrainingArguments
)
SEPARATOR_STRING = '<|endprompt|>'
def customize_tokenizer(tokenizer):
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.truncation_side = 'left'
    tokenizer.add_special_tokens({'sep_token': SEPARATOR_STRING})
```

```python
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
customize_tokenizer(tokenizer)
```

```python
for i in range(data.shape[0]):
    test_pt=tokenizer.decode(data[i,:-1])
    if "bird" in test_pt:
        print(test_pt)
```
